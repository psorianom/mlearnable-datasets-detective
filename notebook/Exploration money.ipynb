```python
import pandas as pd 
from pathlib import Path
import os
from os.path import isfile
```

```python
data_folder = Path(os.getcwd()).parent / 'data' / 'data_ml_money' 
```

```python
data_file = 'all_dabl_analysis.csv'
```

```python
list_file = []
for file in os.listdir(data_folder):
    if isfile(data_folder / file):
        list_file.append(data_folder / file)
```
