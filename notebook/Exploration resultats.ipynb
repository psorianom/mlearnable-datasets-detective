```python
import pandas as pd 
from pathlib import Path
import os
from os.path import isfile
```

```python
data_folder = Path(os.getcwd()).parent / 'data' / 'data_ml_csv' 
```

```python
data_file = 'all_dabl_analysis.csv'
```

```python
list_file = []
for folder in os.listdir(data_folder):
    if not isfile(data_folder / folder):
        for file in os.listdir(data_folder / folder):
            list_file.append(data_folder / folder/ file)
```

```python
file = list_file[0]
df = pd.read_csv(file)
df.head()
```
